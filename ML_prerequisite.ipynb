```python
# Q1. How many multiplications and additions do you need to perform a matrix multiplication 
# between a (n, k) and (k, m) matrix? Explain.
# Ans. n*m*k
# The final answer matrix has the dimensions of nxm, i.e., there are n*m cells. For each cell, we need to multiply k times
# and then add those products. This is repeated for each cell. 


# Q2. Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy.
# Compare the timing of both solutions. Which one is faster? Why?

# using python lists-

import time
n, k, m = 3, 3, 3

A = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]

B = [[5, 2, 3],
     [1, 5, 6],
     [7, 3, 9]]

# ans_arr = [[0]*3]*3
ans_arr = [[0, 0, 0],
           [0, 0, 0],
           [0, 0, 0]]

# print (ans_arr)
start = time.time()

for i in range (n):
    for j in range (m):
        temp = 0
        for l in range (k):
            # ans_arr[i][j] =+ A[i][l] * B[l][j]
            temp = temp + A[i][l] * B[l][j]
            
        ans_arr[i][j] = temp
print ("The multiplication of the two arrays is: ", end = "")
end = time.time()
print(ans_arr)
print("Time elapsed:", end-start)

# time : 0.0012080669403076172
```

```python
# Q2. using numpy
import numpy as np
import time
ans_arr2 = [[0, 0, 0],
           [0, 0, 0],
           [0, 0, 0]]
start = time.time()
ans_arr2 = np.matmul(A, B)
end = time.time()
print (ans_arr2)
print ("Time elapsed:", end-start)

# time : 0.0010137557983398438

# clearly the time is less when we use numpy. 
# It is because numpy carries out operations parallely so it becomes effecient
```

```python
# Q5 grad using jax

import jax 
import jax.numpy as np

def fn(x):
    return x[1]*x[0]**2 + sin(x[0])*x[1]**3

def grad_fn (x):
    return 2*x[0]*x[1] + cos(x[0])*x[1]**3, x[0]**2 + 3*sin(x[0])*x[1]**2

grad_np = jax.grad(fn)
grad_at_x = grad_np([1,1])
print (grad_at_x)
```
